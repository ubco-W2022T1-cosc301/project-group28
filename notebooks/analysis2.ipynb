## Research question/interests

I wish to find out if Center Attack Midfielders are generally better than players in other positions, as CAM are often thought of as the most technically gifted.

```python
# Import libraries including custom library project_functions1 and instantiates a pandas dataframe
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import project_functions1 as pf1
import project_functions2 as pf2

df = pd.read_csv("../data/raw/FIFA22_official_data.csv")

```

```python
# Cleans the pandas dataframe so that we can more effectively use it to answer the research question
clean_df = pf2.df_clean(df)
print(clean_df.head())
print(clean_df.shape)
print(clean_df.columns)

print(df.Position.unique())

df.nunique(axis=0)
```

```python
plt.figure(figsize=(20,15))
chart = sns.boxplot(x='Position', y='Overall', data=clean_df, width=0.5)
sns.despine()
```

Based on this boxplot we can see that the Overall media stat of the CAM position is slightly higher than many other positions. The interquartile range is about the same as the majority of other poistions. While the max Overall stat is around the same as other positions with a slighlty less than average number outliers.

```python
plt.figure(figsize=(40,15))
plt.subplot(231)
plt.scatter(clean_df['Position'],clean_df['Dribbling'])
plt.subplot(232)
plt.scatter(clean_df['Position'],clean_df['Crossing'])
plt.subplot(233)
plt.scatter(clean_df['Position'],clean_df['BallControl'])
plt.subplot(234)
plt.scatter(clean_df['Position'],clean_df['Positioning'])
plt.subplot(235)
plt.scatter(clean_df['Position'],clean_df['Interceptions'])
plt.subplot(236)
plt.scatter(clean_df['Position'],clean_df['ShortPassing'])

plt.show()
```

CAM are considered the "most" technically gifted, however when looking at specific stats like Dribbling which takes a lot of technical skill players in the CAM position seem to be fairly equal with players in other positions even being beaten out by many positions. However, where CAM players seem to be gifted is in there ball control and shortpassing. They also seem to have short comings when it comes to Interceptions. This could be due to their position on the field not being defensive.

Seeing the interception stat I will take a look at the more defensive oriented stats to see if CAM players struggle with defense due to the position on the field or if the people that play this position are less likely to be good defenders.

```python
plt.figure(figsize=(20,15))
chart = sns.boxplot(x='Position', y='StandingTackle', data=clean_df, width=0.5)
sns.despine()
plt.figure(figsize=(20,15))
chart = sns.boxplot(x='Position', y='DefensiveAwareness', data=clean_df, width=0.5)
sns.despine()
plt.figure(figsize=(20,15))
chart = sns.boxplot(x='Position', y='SlidingTackle', data=clean_df, width=0.5)
sns.despine()
```

From these three boxplots we can see that the CAM position is most likely not in a great position on the field to play an integral role in defense. Thus the defensive stats should not be used to determine if a player whos position is CAM is good or not.

### Conclusion

From this EDA we can see that the players in the CAM position maybe very good, they do not outshine other positions players in other roles.
